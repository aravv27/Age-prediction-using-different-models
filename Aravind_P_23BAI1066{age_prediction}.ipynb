```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv(r'Human_Age_Prediction.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
data = data.fillna(0)
data.isnull().sum()
data.head()
```

```python
data['Gender'] = data['Gender'].map({"Male":1,"Female":0})
data.head()
```

```python
x = data.drop("Age (years)",axis = 1)
y = data["Age (years)"]
```

```python
x[['Systolic_BP', 'Diastolic_BP']] = x['Blood Pressure (s/d)'].str.split('/', expand=True)

x['Systolic_BP'] = pd.to_numeric(x['Systolic_BP'])
x['Diastolic_BP'] = pd.to_numeric(x['Diastolic_BP'])
# drop_col = ["Education Level",'Alcohol Consumption','Family History',"Smoking Status","Income Level","Blood Pressure (s/d)"]
# x = x.drop(columns = drop_col)

x.head()
```

```python
x = x.drop(columns=['Gender', 'Blood Pressure (s/d)','Physical Activity Level',
 'Smoking Status',
 'Alcohol Consumption',
 'Diet',
 'Chronic Diseases',
 'Medication Use',
 'Family History',
 'Cognitive Function',
 'Mental Health Status',
 'Sleep Patterns','Education Level',
 'Income Level'])

print(x.columns)
```

```python
label_encoder = LabelEncoder()
string_columns = x.select_dtypes(include=['object']).columns
for column in string_columns:
    x[column] = x[column].astype(str)
    x[column] = label_encoder.fit_transform(x[column])
```

```python
print(x.columns)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 45)
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
# pca = PCA(n_components=5)
# x_train = pca.fit_transform(x_train)
# x_test = pca.fit_transform(x_test)
```

```python
models = {
    'Support Vector Machine': SVC(),
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Linear Regression': LinearRegression(),
    'Random Forest Regressor': RandomForestRegressor(),
    'XGBoost Regressor': XGBRegressor()
}

rf_param_grid = {
    'n_estimators': [50, 100, 200],
    'max_features': ['log2', 'sqrt'],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}

# Hyperparameter tuning for XGBoost
xgb_param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    'subsample': [0.5, 0.75, 1.0]
}

scores = {}

for name, model in models.items():
    if name == 'Random Forest Regressor':
        grid_search = GridSearchCV(model, rf_param_grid, cv=5, scoring='r2', n_jobs=-1)
        grid_search.fit(x_train, y_train)
        best_model = grid_search.best_estimator_
    elif name == 'XGBoost Regressor':
        grid_search = GridSearchCV(model, xgb_param_grid, cv=5, scoring='r2', n_jobs=-1)
        grid_search.fit(x_train, y_train)
        best_model = grid_search.best_estimator_
    else:
        best_model = model
        best_model.fit(x_train, y_train)

    y_pred = best_model.predict(x_test)
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test,y_pred)
    r2 = r2_score(y_test, y_pred)
    scores[name] = {'MSE': mse, 'R2': r2,"MAE":mae}

# Print the evaluation scores
for model_name, score in scores.items():
    print(f"{model_name}: MSE = {score['MSE']:.4f}, R2 = {score['R2']:.4f},MAE = {score['MAE']:.4f}")
```

```python
scores_df = pd.DataFrame.from_dict(scores, orient='index')
print(scores_df)
```
